```python
means = [[0.3,0.3], [0.3, -0.3], [-0.3,-0.3], [-0.3, 0.3]]
sigmas = [0.1, 0.09, 0.04, 0.1]
X, Y = MixGauss(means=means, sigmas=sigmas, n=100)
plt.figure(1)
plt.title("Dataset 1")
plt.scatter(X[:,0], X[:,1], s=40, c=Y, alpha=0.8) 
plt.show()
```

```python
centers = np.array([[0,0], [-0.5, -0.5], [-0.2, 0.3], [0.1, -0.3]])
plt.figure(2)
plt.scatter(X[:,0], X[:,1], s=40, c=Y, alpha=0.8)
plt.scatter(centers[:,0], centers[:, 1], s= 100, c=centers[:,0], marker='^')

plt.show()
```

```python
clusters, centroids = LLoyd(X=X, centers=centers, maxiter=200)
```

```python
plt.figure(3)
plt.scatter(X[:,0], X[:,1], s=40, c=clusters, alpha=0.5)
plt.scatter(centroids[:,0], centroids[:, 1], s= 150, c=[4,5,6,7], marker='^')


plt.show()
```
